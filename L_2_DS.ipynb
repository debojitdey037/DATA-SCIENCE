```python
a = "today "
b = "  weather  "
c = " is GooD  "

a = a.capitalize()
b = b.strip()
c = c.lower()

x = a+b+c;
# Replace
x = x.replace("good","bad")
print(x)
# Find
print(x.find("is"))
# Index
print(x.index("is"))
# Startswith
print(x.startswith("Today"))
# Endswith
print(x.endswith("bad "))

```

```python
s = "A man, a plan, a canal: Panama"
s = s.lower()
new_st = ""

for i in s:
  if (i>='a' and i<='z') or (i>='0' and i<='9'):
    new_st += i

rev = ""

for i in new_st:
  rev = i + rev

if(new_st==rev):
  print("true");
else:
  print("false")

# rev = new_st[: : -1]
```
