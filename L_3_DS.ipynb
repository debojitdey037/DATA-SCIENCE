```python
import numpy as np

matrix = np.random.randint(1,90,size=(5,5))
print(matrix)

print("\n")

# Row wise mean calculation
a = np.mean(matrix,axis = 1);
for i in range (5):
  print("Row:",i+1,a[i])

print("\n")

# Column wise mean calculation
b = np.mean(matrix,axis = 0);
for i in range (5):
  print("column:",i+1,b[i])

print("\n")

# Replacing all even numbers with -1

matrix = np.where(matrix % 2 == 0, -1, matrix)
print(matrix)

print("\n")

#Sort the array row wise

row_sorted = np.sort(matrix, axis=1)
print(row_sorted)

print("\n")

# Sort the array column wise

col_sorted = np.sort(matrix, axis=0)
print(col_sorted)

print("\n")

# Find row and column wise sum separately

row_sum = np.sum(matrix, axis=1)

for i in range(5):
  print("Row:",i+1,row_sum[i])

print("\n")

col_sum = np.sum(matrix, axis=0)

for i in range(5):
  print("column:",i+1,col_sum[i])

print("\n")

# Count the number of -ve's in each column

# count_col = np.sum(matrix < 0, axis=0)
# print(count_col)

count_col = np.sum(np.where(matrix<0,1,0),axis = 0)
print(count_col)



```

```python
stock = np.random.randint(1,100,size=(5,5))
print(stock)

print("\n")

demand = np.random.randint(1,100,size=(5,5))
print(demand)

print("\n")

Remaining_stock = stock - demand
print(Remaining_stock)

print("\n")

Modified_stock = np.where(Remaining_stock < 0, 0, Remaining_stock)
print(Modified_stock)

print("\n")

# Total Shortage
total_shortage = np.sum(np.where(Remaining_stock<0),Remaining_stock * -1, 0)
print(total_shortage)


print("\n")

# Product with zero stock
zero_count = np.sum(np.where(Remaining_stock == 0,1,0), axis = 0)
print(zero_count)

print("\n")

# Sort each row based on remaining stock
sorted_stock = np.sort(Modified_stock, axis=1)
print(sorted_stock)



```
